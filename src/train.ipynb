```python
# 1. Imports
import tensorflow as tf
import numpy as np
import pandas as pd

from ResNet101V2 import ResNet101V2Model
from ConvNeXt import ConvNeXtModel
from DataGenerator import DataGenerator  # Stelle sicher, dass DataGenerator im PYTHONPATH liegt

# Seed setzen
tf.random.set_seed(42)
np.random.seed(42)

# 2. Dataset laden
ukgm = pd.read_csv("/kaggle/input/ukgmdata/dataset.csv")
ukgm['image_path'] = '/kaggle/input/ukgmdata/' + ukgm['image_path'].astype(str)
ukgm['label'] = ukgm['label'].astype(int)
```

```python
# 3. Daten splitten
train_df, val_df, test_df = DataGenerator.split_data(ukgm)

# 4. Generatoren erstellen
train_gen_resnet, val_gen_resnet = DataGenerator.get_generators(train_df, val_df, model_type="resnet", batch_size=32)
train_gen_convnext, val_gen_convnext = DataGenerator.get_generators(train_df, val_df, model_type="convnext", batch_size=32)

# 5. Modelle initialisieren
num_classes = ukgm['label'].nunique()

resnet_model = ResNet101V2Model(input_shape=(224, 224, 3), num_classes=num_classes)
convnext_model = ConvNeXtModel(input_shape=(224, 224, 3), num_classes=num_classes)

# Optional: Model summary anzeigen
#resnet_model.model.summary()
#convnext_model.model.summary()

# 6. Training starten
resnet_model.train(train_gen_resnet, val_gen_resnet, epochs_feature=5, epochs_finetune=25, fine_tune_layers=40)
convnext_model.train(train_gen_convnext, val_gen_convnext, epochs_feature=5, epochs_finetune=10)

# 7. Modelle speichern
#resnet_model.model.save("../models/resnet_model.h5")
#convnext_model.model.save("../models/convnext_model.h5")
```
